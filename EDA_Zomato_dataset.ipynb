#Oberservation:Zomato maximum records and transaction are from India,united states,united kingdom

```python
Rating=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
Rating
```

##Observation
1. when rating is between 4.5 to 4.9------> Excellent
2. when  rating is between 4 to 4.4 ----->Very Good
3. whne ratiin is between 3.5 to 3.9------>Good
4. when rating is betwwen 2.5 to 3.4---->Average
5. when rating is between 0 to 2.4 ----->poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(data=Rating,x='Aggregate rating',y='Rating count')
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(data=Rating,x='Aggregate rating',y='Rating count',hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

Observation: 
1.Not Rated count is very high 
2.Maximun no. of rating is between 2.5 to 3.4

```python
##count plot
sns.countplot(x='Rating color',data=Rating,palette=['blue','red','orange','yellow','green','green'])
```
